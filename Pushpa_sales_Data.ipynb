```python
# Get the total number of rows and columns
rows, columns = sales_data.shape
print("Total rows:", rows)
print("Total columns:", columns)
```

```python
sales_data = sales_data[['OrderID', 'CustomerName', 'Product', 'TotalPrice']]
```

```python
north_orders=sales_data[sales_data['Region'] == 'North']
```

```python
# Filter orders where Quantity is greater than 5
high_quantity_orders = sales_data[sales_data['Quantity'] > 5]

# Display the result
high_quantity_orders
```

```python
# Select orders where Product is "Widget A" or "Gadget B"
selected_orders = sales_data[sales_data['Product'].isin(['Widget A', 'Gadget B'])]

# Display the result
selected_orders
```

```python
# Check for missing values in each column
missing_values = sales_data.isnull().sum()
missing_values
```

```python
# Count null values per column
null_counts = sales_data.isnull().sum()
null_counts
```

```python
# Remove rows with any missing values
sales_data_clean = sales_data.dropna()

#  check the result
sales_data_clean
```

```python
# Remove rows where Quantity or UnitPrice is null
sales_data_clean = sales_data.dropna(subset=['Quantity', 'UnitPrice'])
# check the result
sales_data_clean
```

```python
# Check if there are any duplicate rows in the dataset
duplicate_rows =sales_data.duplicated()
```

```python
sales_data
```

```python
# Remove duplicate rows (keep the first occurrence)
sales_data.drop_duplicates(inplace=True)
```

```python
sales_data
```

```python
# Rename the 'CustomerName' column to 'Customer'
sales_data.rename(columns={'CustomerName': 'Customer'}, inplace=True)
```

```python
sales_data
```

```python
#Rename UnitPrice to PricePerUnit.
sales_data.rename(columns={'UnitPrice': 'PricePerUnit'}, inplace=True)
```

```python
sales_data
```

```python
# Convert the 'Date' column to datetime type
# Convert Date to datetime and display it with time (even if it's just 00:00:00)
sales_data['Date'] = sales_data['Date'].dt.strftime('%d/%m/%Y %H:%M:%S')
```

```python
sales_data
```

```python
# Handling nulls: Replace NaN values in 'Quantity' with 0
sales_data['Quantity'] = sales_data['Quantity'].fillna(0)

# Convert 'Quantity' to integer type
sales_data['Quantity'] = sales_data['Quantity'].astype(int)
```

```python
sales_data
```

```python
# Convert 'TotalPrice' to float
sales_data['TotalPrice'] = sales_data['TotalPrice'].astype(float)
```

```python
sales_data
```

```python
# Filter orders where Category is 'Gadgets' and TotalPrice > 150
filtered_orders = sales_data[(sales_data['Category'] == 'Gadgets') &
 (sales_data['TotalPrice'] > 150)]

# Display the filtered orders
print(filtered_orders)
```

```python
# Sort the data by TotalPrice in descending order
sorted_data = sales_data.sort_values(by='TotalPrice', ascending=False)

# Display the sorted data
print(sorted_data)
```

```python
# Sort the data by TotalPrice in descending order and select the top 10 rows
top_10_orders = sales_data.sort_values(by='TotalPrice', ascending=False).head(10)

# Display the top 10 orders
print(top_10_orders)
```

```python
# Sort the data by Date in descending order and select the last 5 rows (which are the most recent)
last_5_orders = sales_data.sort_values(by='Date', ascending=False).head(5)

# Display the last 5 orders
print(last_5_orders)
```

```python
# Group the data by Region and calculate the total sales per Region
total_sales_per_region = sales_data.groupby('Region')['TotalPrice'].sum()

# Display the result
print(total_sales_per_region)
```

```python
# Group the data by Product and calculate the average TotalPrice per Product
average_total_price_per_product = sales_data.groupby('Product')['TotalPrice'].mean()

# Display the result
print(average_total_price_per_product)
```

```python
# Group the data by Customer and count the number of orders for each customer
orders_per_customer = sales_data.groupby('Customer')['OrderID'].size()

# Display the result
print(orders_per_customer)
```
